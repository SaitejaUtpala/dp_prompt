```python
import numpy as np
from sklearn.model_selection import train_test_split

total = np.arange(len(clean_df))
train_idxs, test_idxs = train_test_split(total, test_size=0.30,  random_state=42)
train_idxs, val_idxs = train_test_split(train_idxs, test_size=0.10,  random_state=42)
print("train len", len(train_idxs))
print("val len", len(val_idxs))
print("test len", len(test_idxs))
```

```python
from attacks.embedding_attack import get_embedding_attack_stats
```

```python
columns_to_run =  columns=[f"madlib_eps={eps}" for eps in madlib_eps_list]
print(columns_to_run)
madlib_attack_results = get_embedding_attack_stats(
    clean_df,
    private_df=private_madlib_df,
    columns_to_run=columns_to_run,
    train_idx_list=train_idxs,
    val_idx_list=val_idxs,
    test_idx_list=test_idxs,
    encode = True
)
```

```python
columns_to_run = [col for col in private_chatgpt_df.columns if "t=" in col ]
print(columns_to_run)
chatgpt_attack_results = get_embedding_attack_stats(
    clean_df,
    private_df=private_chatgpt_df,
    columns_to_run=columns_to_run,
    train_idx_list=train_idxs,
    val_idx_list=val_idxs,
    test_idx_list=test_idxs,
    encode = True
)
```

```python
def structure_f1_scores_results(data):
    f1_scores_results = {
        "static": {"author_id": [], "sentiment_id": []},
        "adaptive": {"author_id": [], "sentiment_id": []},
    }
    for epsilon_key, epsilon_data in data.items():
        for mode in ["static", "adaptive"]:
            for task in ["author_id", "sentiment_id"]:
                f1_score = epsilon_data[mode][task]["F1 Score"]
                f1_scores_results[mode][task].append(f1_score)
    return f1_scores_results
```

```python
import matplotlib.pyplot as plt


chatgpt_results = structure_f1_scores_results(chatgpt_attack_results)
madlib_results = structure_f1_scores_results(madlib_attack_results)

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

for i, mode in enumerate(["static", "adaptive"]):
    axs[i].plot(
        chatgpt_results[mode]["author_id"],
        chatgpt_results[mode]["sentiment_id"],
        "-o",
        label="DP-Prompt (ChatGPT)",
    )
    axs[i].plot(
        madlib_results[mode]["author_id"],
        madlib_results[mode]["sentiment_id"],
        "-x",
        label="Madlib",
    )
    axs[i].set_xlabel("F1 Score (author_id)")
    axs[i].set_ylabel("F1 Score (sentiment_id)")
    axs[i].set_title(f"{mode} attacker")
    axs[i].legend()

plt.tight_layout()
plt.show()
```
